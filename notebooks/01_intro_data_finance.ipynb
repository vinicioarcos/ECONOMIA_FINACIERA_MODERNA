```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from data.generate_data import generate_stock_prices, generate_interest_rates

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
print('Librerías cargadas ✓')
```

## 1. Carga de datos

Vamos a trabajar con precios simulados de 5 activos financieros durante 2 años de trading (≈504 días hábiles). Los datos siguen un **Movimiento Browniano Geométrico**, el modelo estándar de precios en finanzas.

```python
# Cargar datos
prices = generate_stock_prices(n_days=504, n_assets=5)

print('Dimensiones del dataset:', prices.shape)
print('Período:', prices.index[0].date(), '→', prices.index[-1].date())
print('\nPrimeras filas:')
prices.head()
```

## 2. Visualización de precios

El primer paso en cualquier análisis financiero: **ver los datos**.

```python
fig, ax = plt.subplots(figsize=(13, 5))

colors = ['#2563EB', '#DC2626', '#16A34A', '#D97706', '#7C3AED']

for col, color in zip(prices.columns, colors):
    ax.plot(prices.index, prices[col], label=col, color=color, linewidth=1.5)

ax.set_title('Precios de Activos Simulados', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Precio (USD)')
ax.legend(loc='upper left', framealpha=0.9)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Estadísticas descriptivas básicas

```python
summary = pd.DataFrame({
    'Precio inicial': prices.iloc[0].round(2),
    'Precio final':   prices.iloc[-1].round(2),
    'Máximo':         prices.max().round(2),
    'Mínimo':         prices.min().round(2),
    'Retorno total (%)': ((prices.iloc[-1] / prices.iloc[0] - 1) * 100).round(2),
})

print('Resumen de activos:')
summary
```

## 4. ¿Cuál activo fue más rentable? ¿Cuál más volátil?

Rentabilidad y riesgo son las dos dimensiones fundamentales de cualquier activo financiero.

```python
# Retorno total
total_returns = (prices.iloc[-1] / prices.iloc[0] - 1) * 100

# Volatilidad diaria (proxy de riesgo)
daily_returns = prices.pct_change().dropna()
volatility = daily_returns.std() * 100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))

# Retornos
bars1 = ax1.bar(total_returns.index, total_returns.values, color=colors)
ax1.set_title('Retorno Total (%)', fontweight='bold')
ax1.set_ylabel('%')
ax1.axhline(0, color='black', linewidth=0.8)

# Volatilidad
bars2 = ax2.bar(volatility.index, volatility.values, color=colors)
ax2.set_title('Volatilidad Diaria (%)', fontweight='bold')
ax2.set_ylabel('Desviación estándar (%)')

plt.suptitle('Rentabilidad vs Riesgo', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f'Activo más rentable: {total_returns.idxmax()} ({total_returns.max():.1f}%)')
print(f'Activo más volátil:  {volatility.idxmax()} ({volatility.max():.2f}% diario)')
```
